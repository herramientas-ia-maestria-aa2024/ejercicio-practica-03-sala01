```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
from faker import Faker
```

```python
# Crear objeto para datos falsos mediante Faker
fake = Faker()
```

```python
# Crear una lista vacía para almacenar los datos falsos
lista = []
```

```python
# Generar los datos falsos y almacenarlos en la lista creada anteriormente
for x in range(0,1000):
    lista.append({'name': fake.name(),
                  'last_name': fake.last_name(),
                  'email': fake.email(),
                  'address': fake.address()})
```

```python
# Crear un DataFrame de Pandas desde la lista de datos falsos
data = pd.DataFrame(lista)
```

```python
# Mostrar 10 datos aleatorios del dataframe
data.sample(10)
```

```python
# Función para generar la edad de manera aleatoria entre 30 y 40 años
def random_edad(d):
    datos = np.random.randint(30, 41, 1)
    return int(datos)

# Función para generar la provincia de manera aleatoria usando las mencionadas
def random_provincia(d):
    datos = np.random.choice(['Loja', 'Pichincha', 'Guayas', 'Azuay', 'Manabí'])
    return datos
```

```python
# Generar los datos aleatorios de edad y provincia y almacenarlos en el dataframe
data['Edad'] = data['name'].apply(random_edad)
data['Provincia'] = data['name'].apply(random_provincia)
```

```python
# Mostrar 10 datos aleatorios del dataframe
data.sample(10)
```

```python
# Guardar el dataframe como archivo CSV
data.to_csv("midata_grupo_01.csv")
```
